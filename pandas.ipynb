# Data Analysis with Pandas üêº
**Introduction to Pandas**

**Pandas** is a Python library used for data analysis and manipulation. It is built on top of NumPy, so you already know how arrays work. Pandas makes it easier to handle labeled data like tables (rows and columns).

**Pandas mainly has three data structures:**

- **Series** ‚Äì 1D labeled array (like a single column in Excel).

- **DataFrame** ‚Äì 2D labeled table (like Excel sheet with rows and columns).

- **Panel** ‚Äì 3D labeled data (less used today, replaced by MultiIndex DataFrames)

We‚Äôll focus mostly on Series and DataFrame because Panel is rarely used in modern data analysis.

## **Pandas Series üìä**

A Series is a one-dimensional array with labels (index).

```python
# Example: Student marks in Math
math_marks = pd.Series([85, 90, 78, 92, 88], index=['Ali', 'Sara', 'Zia', 'Aisha', 'Bilal'])
print(math_marks)
```

**Concepts:**

`math_marks.values` ‚Üí gives the raw data [85, 90, 78, 92, 88]

`math_marks.index` ‚Üí gives the labels ['Ali', 'Sara', 'Zia', 'Aisha', 'Bilal']

**Access Elements By Label(index)**

```python
print(math_marks['Zia'])
```

**For Example** : Teacher wants to give 2 number extra to every student .

```python
print(math_marks)
print("-"*30)
#add  2 number
math_marks += 2 
print(math_marks)
```

Teacher wants **Boolean Filtering**  and he wants to know , students who obtain 90 plus marks .

```python
high_score = math_marks[math_marks>90]
print(high_score)
```

## **Pandas DataFrame üóÇÔ∏è**

A DataFrame is like a spreadsheet or SQL table. It has rows (index) and columns.

```python
# Example : Student scores in multiple subjects.

data = {
    'Math': [85, 90, 78, 92, 88],
    'Physics': [80, 85, 75, 95, 89],
    'Chemistry': [82, 88, 79, 91, 87]
}

students = pd.DataFrame(data,index=['Ali', 'Sara', 'Zia', 'Aisha', 'Bilal'])
print(students)
```

**Accessing DataFrame Data**

```python
# Accessing a single column
print(students["Math"])
print("")

# Accessing a multiple columns
print(students[['Math','Chemistry']])
```

```python
#Access a row by index 
print(students.loc['Zia'])

print("-"*30)

#Access a row by position 
print(students.iloc[1]) # Sara is present on index 1
```

**Basic DataFrame Operations**

```python
#Add a new column 
students['Biology']=[85, 87, 80, 90, 86]
print(students)
```

```python
#Calculate mean of marks for each student 
students['Average'] = students.mean(axis=1) #axis 1 = row wise mean 
print(students)
```

```python
#sort students by math marks
print(students.sort_values(by='Math' , ascending=False))
```

**Basic Functionality**

```python
#View first 3 rows 
print(students.head(3))
```

```python
#view last 2 rows 
print(students.tail(2))
```

```python
#Describe statistical summary 
print(students.describe())
```

```python
#check info about DataFrame
print(students.info())
```

```python
#check columns name 
print(students.columns)
```

### **Iteration over DataFrame üîÑ**

You can loop through rows, columns, or items.

**Iterate over Columns**

```python
for col_name, col_data in students.items():
    print(f"Column: {col_name}")
    print(col_data)
    print("---")

```

## Text, Date, and Time Handling in Pandas üïíüìù

**1Ô∏è‚É£ Working with Text Data üìù**

Pandas allows easy handling of text (strings) using .str accessor.

**Example:** Student Names

```python
std = pd.DataFrame({
    "Name" : ["Ali",'Anum'],
    "Math" : [20,30]
})
print(std)
```

```python
#convert names to lower case 
std["Lower_case_name"] = std['Name'].str.lower()
print(std)
```

```python
#check which name starts with "A"
std['Starts_with_A'] = std['Name'].str.startswith('A')
print(std['Starts_with_A'])
print(std)
```

```python
#get length of each name 
std['lenght_of_name'] = std['Name'].str.len()
print(std)
```

**Common Text Operations**

- **str.lower() / .str.upper()** ‚Üí Convert case

- **str.startswith('A') / .str.endswith('l')** ‚Üí Check string patterns

- **str.contains('a')** ‚Üí Filter text

- **str.replace('a','@')** ‚Üí Replace characters

## **2Ô∏è‚É£ Working with Date and Time üïí**

Pandas has powerful datetime functions using pd.to_datetime().

Example: Exam Dates

```python
students = pd.DataFrame({
    'Name': ['Ali', 'Sara', 'Zia', 'Aisha', 'Bilal'],
    'Math': [85, 90, 78, 92, 88],
    'Exam_Date': ['2026-01-01', '2026-01-03', '2026-01-05', '2026-01-07', '2026-01-09']
})
```

```python
print(students[['Name','Exam_Date']])
```

```python
print(students.info())
```

```python
students['Exam_Date'] = pd.to_datetime(students['Exam_Date'])
print(students.info())
```

```python
print(students)
```

```python
# Extract year, month, day
students['Year'] = students['Exam_Date'].dt.year
students['Month'] = students['Exam_Date'].dt.month
students['Day'] = students['Exam_Date'].dt.day
print(students)

# Day of the week
students['Weekday'] = students['Exam_Date'].dt.day_name()
print(students)
```

Difference between dates . (har paper ka dermiyan gap kitna tha )

```python
# Exam preparation days before next exam
students['Next_Exam_Date'] = students['Exam_Date'].shift(-1)
students['Days_to_Next'] = students['Next_Exam_Date'] - students['Exam_Date']
print(students)
```

### **3Ô∏è‚É£ Categorical Data üè∑Ô∏è**

Categorical data is used for non-numeric data with finite categories (e.g., grades, levels).

Example: Grades

```python
students = pd.DataFrame({
    'Name': ['Ali', 'Sara', 'Zia', 'Aisha', 'Bilal'],
    'Math': [85, 90, 78, 92, 88],
    'Exam_Date': ['2026-01-01', '2026-01-03', '2026-01-05', '2026-01-07', '2026-01-09']
})
students['Grade'] = pd.cut(students['Math'], bins=[0, 79, 89, 100], labels=['C', 'B', 'A'])
print(students)
```

**Why Categorical Data is useful:**

1. Saves memory for large datasets

2. Easier to group and analyze data

### ‚úÖ **Summary**

1.  **Text Data**: Use .str for string operations (lowercase, length, startswith, contains)
    
2.  **Date/Time Data**: Use pd.to\_datetime() and .dt to extract year, month, day, weekday
    
3.  **Timedelta**: Difference between dates (date2 - date1)
    
4.  **Categorical Data**: Use pd.cut() or astype('category') for discrete categories
